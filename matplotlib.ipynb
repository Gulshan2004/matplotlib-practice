SUBPLOTS

IN THIS SECTION WE WILL UNDERSTAND HOW WE CAN PLOT MULTIPLE PLOTS AT THE SAME TIME USING SUBPLOTS

```python
# A diff way of plotting graphs
batter.head()
```

```python
plt.figure(figsize=(15,6))
plt.scatter(batter['avg'],batter['strike_rate'])
plt.title('something')
plt.xlabel('avg')
plt.ylabel('Strike rate')
plt.show()
```

```python

fig, ax=plt.subplots(figsize=(15,6)) #here fig represnts figure and ax represents axis
ax.scatter(batter['avg'],batter['strike_rate'],color='red')
#with the help of above single code we can plot any type of plot
ax.set_title('something')# this is the syntax to write The Title of our plot
ax.set_xlabel('avg')#in place of plt.xlabel()
ax.set_ylabel('strike Rate')# in place of plt.ylabel()
#fig.show()#in place of plt.show()
```

```python
fig,ax=plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(10,6))
#sharex parameter autamatically shares the values of x at the bottom 

ax[0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[1].scatter(batter['avg'],batter['runs'])

ax[0].set_title('avg vs strike_rate')
ax[0].set_ylabel('strike_rate')

ax[1].set_title('avg vs runs')
ax[1].set_ylabel('runs')
ax[1].set_xlabel('Avg')
```

```python
#another way of plotting 'THIS IS PREFFERED'
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,10))
ax[0,0].scatter(batter['avg'],batter['strike_rate'],color='red')
ax[0,1].scatter(batter['avg'],batter['runs'])
ax[1,0].hist(batter['avg'])
ax[1,1].hist(batter['runs'])
```

```python
fig=plt.figure()

ax1=fig.add_subplot(2,1,1)#2 rows ,1 column , 1st plot
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')

ax2=fig.add_subplot(2,1,2)#2 rows ,1 column , 2nd plot
ax2.hist(batter['runs'])
```

```python
#this is the another method of plotting "NOT PREFFERD"
fig=plt.figure()

ax1=fig.add_subplot(2,2,1)#2 rows ,2 column , 1st plot
ax1.scatter(batter['avg'],batter['strike_rate'],color='red')

ax2=fig.add_subplot(2,2,2)#2 rows ,2 column , 2nd plot
ax2.hist(batter['runs'])

ax3=fig.add_subplot(2,2,3)#2 rows ,2 column , 3rd plot
ax3.hist(batter['avg'])
```

3D LINE PLOT

```python
x=[0,1,5,25]
y=[0,10,13,0]
z=[0,13,20,5]
#x, y, z represent positions in 3D space
#Each index (x[i], y[i], z[i]) is a point in 3D
fig=plt.figure()# Creates a new figure window to plot in.Like a blank canvas for your plot.
ax=plt.subplot(projection='3d')
'''plt.subplot() creates a subplot (axes object).
projection='3d' tells Matplotlib to make it a 3D plot, not 2D.
Now ax can handle 3D plotting functions like plot3D, scatter3D.'''
ax.scatter3D(x,y,z,s=[100,100,100,100])
''' This draws the 3D scatter plot (individual dots):
    x, y, z: Coordinates of points
    s: Size of the dots. All 4 dots will be size 100.'''
#scatter3D() is used when you want to plot points in 3D space.
ax.plot3D(x,y,z,color='red')#This draws a 3D line that connects all the points:
# plot3D() is used to draw a line in 3D space
```

3D SURFACE PLOTS

```python
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
# np.linspace() is a NumPy function used to create
#a sequence of evenly spaced numbers between a start and end value.
```

```python
xx,yy=np.meshgrid(x,y)
yy.shape 
#meshgrid() creates a grid of coordinates from two 1D arrays â€” basically,
#it helps you create a 2D or 3D "mesh" of points where each combination of X and Y exists.
```

```python
z=xx**2+yy**2
z.shape
```

```python
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
'''Note:- Here we cannot use colorbar directly that's 
why we have to store our graph in a varible in an order to display the colorbar'''
```

```python
#3d surface plot of sin and cos
'''z=np.sin(xx)+np.cos(yy)
fig=plt.figure(figsize=(12,8))
ax=plt.subplot(projection='3d')
p=ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)'''
```

```python
delivery=pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
#now this is the dataframe on which we will work on 
#ye data frame hame bata ra hai ki ek batsman ne ajj tkk total matches mai over ki kis ball 
#pe sixer mara hai provided ki over ki ball six ko exceed na kare 
temp_df=delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
#we have learnt pivot table in pandas
'''
 syntax:-   df.pivot_table(index='Name', columns='Subject', values='Marks', aggfunc='mean')
 
index   : Jo row-wise group banega (yaha: student name)
columns : Jo column banenge (yaha: subjects)
values	: Jiska hum result nikalna chahte (yaha: marks)
aggfunc : Kya operation karna hai? (e.g., mean, sum, count, etc.)
 '''
grid=temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```
